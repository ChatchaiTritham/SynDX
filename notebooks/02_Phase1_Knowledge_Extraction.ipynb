## 2. TiTrATE Framework Formalization

### 2.1 Initialize TiTrATE Formalizer

The **TiTrATE** framework structures vestibular assessment into 5 components:
- **Ti**ming: Onset pattern (acute, episodic, chronic)
- **Tr**igger: Precipitating factors (head movement, position change)
- **A**ssessment: Associated symptoms (hearing loss, headache, diplopia)
- **T**argeted **E**xam: Physical findings (nystagmus, head impulse test, Dix-Hallpike)

```python
# Initialize TiTrATE formalizer
titrate = TiTrATEFormalizer()

# Get formalized rules
rules = titrate.get_all_rules()

print(f"Total TiTrATE rules extracted: {len(rules)}")
print(f"\nRule categories: {list(rules.keys())}")
```

### 2.2 Visualize TiTrATE Components

```python
# Extract rule counts by category
rule_counts = {category: len(items) for category, items in rules.items()}

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Bar chart
categories = list(rule_counts.keys())
counts = list(rule_counts.values())
colors = sns.color_palette("husl", len(categories))

ax1.bar(categories, counts, color=colors, alpha=0.7, edgecolor='black')
ax1.set_xlabel('TiTrATE Component', fontsize=12, fontweight='bold')
ax1.set_ylabel('Number of Rules', fontsize=12, fontweight='bold')
ax1.set_title('TiTrATE Framework Rule Distribution', fontsize=14, fontweight='bold')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, (cat, cnt) in enumerate(zip(categories, counts)):
    ax1.text(i, cnt + 0.5, str(cnt), ha='center', fontweight='bold')

# Pie chart
ax2.pie(counts, labels=categories, autopct='%1.1f%%', colors=colors, 
        startangle=90, textprops={'fontsize': 10, 'fontweight': 'bold'})
ax2.set_title('TiTrATE Component Proportions', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\nðŸ“Š Rule Distribution:")
for cat, cnt in rule_counts.items():
    print(f"  {cat}: {cnt} rules")
```

### 2.3 Example: Extract Diagnosis-Specific Rules

```python
# Example: Get rules for BPPV diagnosis
diagnosis = "BPPV"
bppv_rules = titrate.get_diagnosis_rules(diagnosis)

print(f"\nðŸ” TiTrATE Rules for {diagnosis}:")
print("=" * 80)

for component, rules_list in bppv_rules.items():
    print(f"\n{component.upper()}:")
    for rule in rules_list:
        print(f"  â€¢ {rule}")

# Visualize BPPV rule pattern
bppv_counts = {comp: len(rules_list) for comp, rules_list in bppv_rules.items()}

plt.figure(figsize=(10, 6))
plt.barh(list(bppv_counts.keys()), list(bppv_counts.values()), 
         color='coral', alpha=0.7, edgecolor='black')
plt.xlabel('Number of Rules', fontsize=12, fontweight='bold')
plt.ylabel('TiTrATE Component', fontsize=12, fontweight='bold')
plt.title(f'TiTrATE Rule Pattern for {diagnosis}', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# Add value labels
for i, (comp, cnt) in enumerate(bppv_counts.items()):
    plt.text(cnt + 0.1, i, str(cnt), va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 3. Archetype Generation

### 3.1 Initialize Archetype Generator

```python
# Initialize archetype generator
generator = ArchetypeGenerator(random_state=42)

# Generate small sample for demonstration
n_samples = 1000
archetypes_df = generator.generate_archetypes(n_samples=n_samples)

print(f"âœ“ Generated {len(archetypes_df)} archetypes")
print(f"\nðŸ“‹ Feature dimensions: {archetypes_df.shape[1]} features")
print(f"\nFirst 5 archetypes:")
print(archetypes_df.head())
```

### 3.2 Analyze Diagnosis Distribution

```python
# Diagnosis distribution
diagnosis_counts = archetypes_df['diagnosis'].value_counts()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart
diagnosis_counts.plot(kind='bar', ax=ax1, color='skyblue', alpha=0.7, edgecolor='black')
ax1.set_xlabel('Diagnosis', fontsize=12, fontweight='bold')
ax1.set_ylabel('Count', fontsize=12, fontweight='bold')
ax1.set_title('Diagnosis Distribution in Generated Archetypes', fontsize=14, fontweight='bold')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(axis='y', alpha=0.3)

# Add percentage labels
for i, (diag, count) in enumerate(diagnosis_counts.items()):
    percentage = (count / n_samples) * 100
    ax1.text(i, count + 10, f'{percentage:.1f}%', ha='center', fontweight='bold')

# Pie chart
colors = sns.color_palette("Set2", len(diagnosis_counts))
ax2.pie(diagnosis_counts.values, labels=diagnosis_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90, textprops={'fontsize': 9, 'fontweight': 'bold'})
ax2.set_title('Diagnosis Proportions', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nðŸ“Š Diagnosis Statistics:")
for diag, count in diagnosis_counts.items():
    print(f"  {diag}: {count} ({count/n_samples*100:.2f}%)")
```

### 3.3 Analyze Demographics

```python
# Age and gender distribution
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Age distribution
axes[0, 0].hist(archetypes_df['age'], bins=30, color='lightcoral', 
                alpha=0.7, edgecolor='black')
axes[0, 0].set_xlabel('Age', fontsize=11, fontweight='bold')
axes[0, 0].set_ylabel('Frequency', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Age Distribution', fontsize=12, fontweight='bold')
axes[0, 0].axvline(archetypes_df['age'].mean(), color='red', 
                    linestyle='--', linewidth=2, label=f'Mean: {archetypes_df["age"].mean():.1f}')
axes[0, 0].legend()
axes[0, 0].grid(axis='y', alpha=0.3)

# Gender distribution
gender_counts = archetypes_df['gender'].value_counts()
axes[0, 1].pie(gender_counts.values, labels=gender_counts.index, 
               autopct='%1.1f%%', colors=['lightblue', 'pink'],
               textprops={'fontsize': 11, 'fontweight': 'bold'})
axes[0, 1].set_title('Gender Distribution', fontsize=12, fontweight='bold')

# Age by diagnosis
diagnosis_order = diagnosis_counts.index
age_by_diagnosis = [archetypes_df[archetypes_df['diagnosis'] == diag]['age'].values 
                    for diag in diagnosis_order]
bp = axes[1, 0].boxplot(age_by_diagnosis, labels=diagnosis_order, patch_artist=True)
for patch, color in zip(bp['boxes'], sns.color_palette("Set2", len(diagnosis_order))):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1, 0].set_xlabel('Diagnosis', fontsize=11, fontweight='bold')
axes[1, 0].set_ylabel('Age', fontsize=11, fontweight='bold')
axes[1, 0].set_title('Age Distribution by Diagnosis', fontsize=12, fontweight='bold')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(axis='y', alpha=0.3)

# Symptom duration distribution
if 'symptom_duration_days' in archetypes_df.columns:
    axes[1, 1].hist(archetypes_df['symptom_duration_days'], bins=30, 
                    color='lightgreen', alpha=0.7, edgecolor='black')
    axes[1, 1].set_xlabel('Duration (days)', fontsize=11, fontweight='bold')
    axes[1, 1].set_ylabel('Frequency', fontsize=11, fontweight='bold')
    axes[1, 1].set_title('Symptom Duration Distribution', fontsize=12, fontweight='bold')
    axes[1, 1].set_xscale('log')
    axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\nðŸ“Š Demographic Statistics:")
print(f"  Age: {archetypes_df['age'].mean():.1f} Â± {archetypes_df['age'].std():.1f} years")
print(f"  Age range: {archetypes_df['age'].min():.0f} - {archetypes_df['age'].max():.0f} years")
print(f"  Gender: {gender_counts.to_dict()}")
```

### 3.4 Analyze TiTrATE Features

```python
# Extract TiTrATE feature columns
timing_cols = [col for col in archetypes_df.columns if 'timing_' in col]
trigger_cols = [col for col in archetypes_df.columns if 'trigger_' in col]
assessment_cols = [col for col in archetypes_df.columns if 'assessment_' in col or 'associated_' in col]
exam_cols = [col for col in archetypes_df.columns if 'exam_' in col or 'nystagmus_' in col or 'hit_' in col]

print(f"\nðŸ” TiTrATE Feature Breakdown:")
print(f"  Timing features: {len(timing_cols)}")
print(f"  Trigger features: {len(trigger_cols)}")
print(f"  Assessment features: {len(assessment_cols)}")
print(f"  Exam features: {len(exam_cols)}")

# Visualize feature group sizes
feature_groups = {
    'Timing': len(timing_cols),
    'Trigger': len(trigger_cols),
    'Assessment': len(assessment_cols),
    'Exam': len(exam_cols)
}

plt.figure(figsize=(10, 6))
plt.bar(feature_groups.keys(), feature_groups.values(), 
        color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'], 
        alpha=0.7, edgecolor='black')
plt.xlabel('TiTrATE Component', fontsize=12, fontweight='bold')
plt.ylabel('Number of Features', fontsize=12, fontweight='bold')
plt.title('Feature Count by TiTrATE Component', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

for i, (group, count) in enumerate(feature_groups.items()):
    plt.text(i, count + 0.5, str(count), ha='center', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()
```

### 3.5 Feature Correlation Heatmap

```python
# Select numeric features for correlation analysis
numeric_cols = archetypes_df.select_dtypes(include=[np.number]).columns[:20]  # First 20 for visibility
correlation_matrix = archetypes_df[numeric_cols].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Heatmap (First 20 Features)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 4. Healthcare Standards Mapping

### 4.1 FHIR Resource Mapping

```python
# Initialize standards mapper
mapper = StandardsMapper()

# Example patient for FHIR mapping
example_patient = archetypes_df.iloc[0].to_dict()

# Generate FHIR resources
fhir_resources = mapper.to_fhir(example_patient)

print("\nðŸ¥ FHIR Resources Generated:")
print("=" * 80)
for resource_type, resource_data in fhir_resources.items():
    print(f"\n{resource_type.upper()}:")
    if isinstance(resource_data, dict):
        for key, value in list(resource_data.items())[:5]:  # Show first 5 fields
            print(f"  {key}: {value}")
        if len(resource_data) > 5:
            print(f"  ... ({len(resource_data) - 5} more fields)")

# Visualize FHIR resource types
resource_types = list(fhir_resources.keys())
resource_counts = [1] * len(resource_types)  # Each type appears once

plt.figure(figsize=(10, 6))
plt.barh(resource_types, resource_counts, color='mediumseagreen', alpha=0.7, edgecolor='black')
plt.xlabel('Count', fontsize=12, fontweight='bold')
plt.ylabel('FHIR Resource Type', fontsize=12, fontweight='bold')
plt.title('FHIR Resources Generated per Patient', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

### 4.2 SNOMED CT Coding

```python
# Map diagnosis to SNOMED CT codes
diagnoses = archetypes_df['diagnosis'].unique()
snomed_mappings = {}

for diag in diagnoses:
    snomed_code = mapper.get_snomed_code(diag)
    if snomed_code:
        snomed_mappings[diag] = snomed_code

print("\nðŸ”¬ SNOMED CT Mappings:")
print("=" * 80)
for diag, code in snomed_mappings.items():
    print(f"  {diag}: {code}")

# Create visualization
if snomed_mappings:
    fig, ax = plt.subplots(figsize=(12, 6))
    
    diagnoses_list = list(snomed_mappings.keys())
    codes_list = [snomed_mappings[d] for d in diagnoses_list]
    
    y_pos = np.arange(len(diagnoses_list))
    colors = sns.color_palette("viridis", len(diagnoses_list))
    
    bars = ax.barh(y_pos, [1]*len(diagnoses_list), color=colors, alpha=0.7, edgecolor='black')
    ax.set_yticks(y_pos)
    ax.set_yticklabels([f"{d}\n({c})" for d, c in zip(diagnoses_list, codes_list)])
    ax.set_xlabel('', fontsize=12, fontweight='bold')
    ax.set_title('Diagnosis to SNOMED CT Code Mapping', fontsize=14, fontweight='bold')
    ax.set_xlim([0, 1.2])
    ax.set_xticks([])
    
    plt.tight_layout()
    plt.show()
```

### 4.3 LOINC Code Mapping

```python
# Map common vestibular tests to LOINC codes
vestibular_tests = [
    'head_impulse_test',
    'dix_hallpike_test',
    'nystagmus_direction',
    'hearing_test',
    'caloric_test'
]

loinc_mappings = {}
for test in vestibular_tests:
    loinc_code = mapper.get_loinc_code(test)
    if loinc_code:
        loinc_mappings[test] = loinc_code

print("\nðŸ§ª LOINC Test Code Mappings:")
print("=" * 80)
for test, code in loinc_mappings.items():
    print(f"  {test}: {code}")

# Visualize LOINC mappings
if loinc_mappings:
    plt.figure(figsize=(12, 6))
    tests = [t.replace('_', ' ').title() for t in loinc_mappings.keys()]
    codes = list(loinc_mappings.values())
    
    plt.barh(tests, [1]*len(tests), color='coral', alpha=0.7, edgecolor='black')
    
    # Add LOINC codes as text
    for i, (test, code) in enumerate(zip(tests, codes)):
        plt.text(0.5, i, code, ha='center', va='center', 
                fontweight='bold', fontsize=10, color='white',
                bbox=dict(boxstyle='round', facecolor='darkblue', alpha=0.8))
    
    plt.xlabel('', fontsize=12)
    plt.ylabel('Vestibular Test', fontsize=12, fontweight='bold')
    plt.title('Vestibular Tests with LOINC Codes', fontsize=14, fontweight='bold')
    plt.xlim([0, 1.2])
    plt.xticks([])
    plt.tight_layout()
    plt.show()
```

## 5. Data Quality Assessment

```python
# Data quality metrics
quality_metrics = {
    'Total Records': len(archetypes_df),
    'Total Features': archetypes_df.shape[1],
    'Missing Values': archetypes_df.isnull().sum().sum(),
    'Duplicate Records': archetypes_df.duplicated().sum(),
    'Completeness (%)': (1 - archetypes_df.isnull().sum().sum() / (archetypes_df.shape[0] * archetypes_df.shape[1])) * 100
}

print("\nâœ“ Data Quality Assessment:")
print("=" * 80)
for metric, value in quality_metrics.items():
    if isinstance(value, float):
        print(f"  {metric}: {value:.2f}")
    else:
        print(f"  {metric}: {value}")

# Visualize data quality
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Quality metrics bar chart
display_metrics = {k: v for k, v in quality_metrics.items() if k not in ['Completeness (%)']}
ax1.bar(display_metrics.keys(), display_metrics.values(), 
        color=['green', 'blue', 'orange', 'red'], alpha=0.7, edgecolor='black')
ax1.set_ylabel('Count', fontsize=12, fontweight='bold')
ax1.set_title('Data Quality Metrics', fontsize=14, fontweight='bold')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(axis='y', alpha=0.3)

# Completeness gauge
completeness = quality_metrics['Completeness (%)']
ax2.barh(['Completeness'], [completeness], color='limegreen', alpha=0.7, edgecolor='black')
ax2.barh(['Completeness'], [100-completeness], left=[completeness], 
         color='lightgray', alpha=0.3, edgecolor='black')
ax2.set_xlim([0, 100])
ax2.set_xlabel('Percentage (%)', fontsize=12, fontweight='bold')
ax2.set_title(f'Data Completeness: {completeness:.2f}%', fontsize=14, fontweight='bold')
ax2.text(completeness/2, 0, f'{completeness:.1f}%', 
         ha='center', va='center', fontweight='bold', fontsize=14, color='white')

plt.tight_layout()
plt.show()
```
